```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35]

# Create a bar chart using Matplotlib
plt.bar(categories, values, color='skyblue')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart using Matplotlib')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Create a line graph
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add labels to the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a title to the graph
plt.title('Line Graph')

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating some example data
np.random.seed(42)
data = [np.random.normal(0, 1, 100),
        np.random.normal(0, 2, 100),
        np.random.normal(0, 1.5, 100),
        np.random.normal(0, 3, 100)]

# Creating a box plot
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'])
plt.title('Box Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create a pie chart
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])

# Add a title
plt.title('Popularity of Programming Languages')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(30) * 10  # Random values for the x-axis
y = 3 * x + np.random.randn(30) * 2  # Linear relationship with some noise for the y-axis

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot Example')

# Display the plot
plt.show()
```
