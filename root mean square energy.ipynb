```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
a = "E:\\audio processing\\audio files\\sample-6s.mp3"
b = "E:\\audio processing\\audio files\\white-noise-8117.mp3"
c = "E:\\audio processing\\audio files\\news-time-intro-13620.mp3"
```

```python
ipd.Audio(a)
ipd.Audio(b)
ipd.Audio(c)
```

```python
audio1 ,_ = librosa.load(a)
audio2 ,_ = librosa.load(b)
audio3 ,_ =librosa.load(c)
```

```python
import librosa.display
import matplotlib.pyplot as plt

FRAME_LENGTH = 1024
HOP_LENGTH = 512

# Assuming you have already loaded the audio data and sampling rate
# using librosa.load, as in your previous code snippet.

# Calculate RMS features
rms_audio1 = librosa.feature.rms(y=audio1, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
rms_audio2 = librosa.feature.rms(y=audio2, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
rms_audio3 = librosa.feature.rms(y=audio3, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)

# Get frame times
frames_audio1 = librosa.frames_to_time(range(len(rms_audio1[0])), hop_length=HOP_LENGTH)
frames_audio2 = librosa.frames_to_time(range(len(rms_audio2[0])), hop_length=HOP_LENGTH)
frames_audio3 = librosa.frames_to_time(range(len(rms_audio3[0])), hop_length=HOP_LENGTH)

# Visualize the RMS features
plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
librosa.display.waveshow(audio1, alpha=0.5)
plt.plot(frames_audio1, rms_audio1[0], color='r', alpha=0.75)
plt.title('RMS Feature - audio1')

plt.subplot(3, 1, 2)
librosa.display.waveshow(audio2, alpha=0.5)
plt.plot(frames_audio2, rms_audio2[0], color='r', alpha=0.75)
plt.title('RMS Feature - audio2')

plt.subplot(3, 1, 3)
librosa.display.waveshow(audio3, alpha=0.5)
plt.plot(frames_audio3, rms_audio3[0], color='r', alpha=0.75)
plt.title('RMS Feature - audio3')

plt.show()
print(rms_audio1.shape)
```

```python
import numpy as np
def rms(signal, frame_length , hop_length):
    rms = []
    for i in range(0 , len(signal) , hop_length):
        rms_current_frame = np.sqrt(np.sum(signal[i:i+frame_length]**2)/frame_length)
        
        rms.append(rms_current_frame)
    return np.array(rms)
        
    
```

```python
rms_audio1 = rms(audio1, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
rms_audio2 = rms(audio2, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
rms_audio3 = rms(audio3, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
```

```python
zcr_audio1 = librosa.feature.zero_crossing_rate(y=audio1, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
zcr_audio2 = librosa.feature.zero_crossing_rate(y=audio2, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)
zcr_audio3 = librosa.feature.zero_crossing_rate(y=audio3, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)

frames_audio1 = librosa.frames_to_time(range(len(zcr_audio1[0])), hop_length=HOP_LENGTH)
frames_audio2 = librosa.frames_to_time(range(len(zcr_audio2[0])), hop_length=HOP_LENGTH)
frames_audio3 = librosa.frames_to_time(range(len(zcr_audio3[0])), hop_length=HOP_LENGTH)

# Visualize the RMS features
plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
librosa.display.waveshow(audio1, alpha=0.5)
plt.plot(frames_audio1, zcr_audio1[0], color='r', alpha=0.75)
plt.title('RMS Feature - audio1')

plt.subplot(3, 1, 2)
librosa.display.waveshow(audio2, alpha=0.5)
plt.plot(frames_audio2, zcr_audio2[0], color='r', alpha=0.75)
plt.title('RMS Feature - audio2')

plt.subplot(3, 1, 3)
librosa.display.waveshow(audio3, alpha=0.5)
plt.plot(frames_audio3, zcr_audio3[0], color='r', alpha=0.75)
plt.title('RMS Feature - audio3')

plt.show()
print(rms_audio1.shape)

```
